# Procesamiento de datos

```python
import numpy as np
import pandas  as pd
```

```python
# se procede a realizar la limpieza de datoss
df = pd.read_csv('../data/raw/netflix_titles.csv')
# copia de la data para no alterar el original
df_cleaned = df.copy()
```

### Tratamieto de datos nulos

```python
# datos NaN de director, Cast, Country  se reemplazan por 'Desconocido'
df_cleaned['director'].fillna('Unknown', inplace=True)
df_cleaned['cast'].fillna('Unknown', inplace=True)
df_cleaned['country'].fillna('Unknown', inplace=True)

# datos NaN de date_added se reemplazan por la fecha de lanzamiento
df_cleaned['date_added'].fillna(df_cleaned['release_year'], inplace=True)
```

###  Nuevas columnas

```python
# Datatime de date_added
df_cleaned['year_added'] = pd.to_datetime(df_cleaned['date_added'], errors='coerce').dt.year
df_cleaned['month_added'] = pd.to_datetime(df_cleaned['date_added'],errors='coerce').dt.month.map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})

df_cleaned.head()
```

```python
df_cleaned.isnull().sum()

# Se verifican los datos que quedaron nulos en las nuevas columnas por el formato de procedencia

```

```python
df_cleaned[df_cleaned['year_added'].isnull()]

# Se observa que los datos nulos en year_added y month_added corresponden a registros con formato de date_added no reconocido, posiblemente por espacios adicionales o errores tipográficos. 

#Ademas se identifica que hay registros solo con el año de fecha
```

```python
# Se identifican los registros que solo contienen el año en date_added. y se puede observar que tienen el mismo que el de release_year. por lo cual se utilizara estos valores para completar year_added

df_cleaned[df_cleaned['date_added'].astype(str).str.match(r'^\d{4}$')]
```

```python
# para los registros con posibles espacios adicionales se procede a limpiar los espacios y volver a convertir a datetime
df_cleaned['date_added'] = df_cleaned['date_added'].astype(str).str.strip()
df_cleaned['year_added'] = pd.to_datetime(df_cleaned['date_added'], errors='coerce').dt.year
df_cleaned['month_added'] = pd.to_datetime(df_cleaned['date_added'],errors='coerce').dt.month.map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})

df_cleaned.loc[df_cleaned['year_added'].isnull()]
```

```python
# Se procede a completar los valores de year_added con release_year para los registros que solo contienen el año en date_added

mask = (
    df_cleaned['year_added'].isna() &
    df_cleaned['date_added'].astype(str).str.match(r'^\d{4}$')
)

df_cleaned.loc[mask, 'year_added'] = df_cleaned.loc[mask, 'release_year']

# Finalmente, se rellenan los valores restantes nulos en  month_added con 'Unknown'
df_cleaned['month_added'] = df_cleaned['month_added'].fillna('Unknown')

df_cleaned.isnull().sum()
```
