# Procesamiento de datos

```python
import numpy as np
import pandas  as pd
```

```python
# se procede a realizar la limpieza de datoss
df = pd.read_csv('../data/raw/netflix_titles.csv')
# copia de la data para no alterar el original
df_cleaned = df.copy()
```

## Tratamieto de datos nulos

```python
# datos NaN de director, Cast, Country  se reemplazan por 'Desconocido'
df_cleaned['director'].fillna('Unknown', inplace=True)
df_cleaned['cast'].fillna('Unknown', inplace=True)
df_cleaned['country'].fillna('Unknown', inplace=True)

# datos NaN de date_added se reemplazan por la fecha de lanzamiento
df_cleaned['date_added'].fillna(df_cleaned['release_year'], inplace=True)
```
